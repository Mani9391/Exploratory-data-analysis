```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/train_and_test2.csv')
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

```python
df.info()
```

```python
sns.countplot(data = df, x = 'Survived', hue = 'Sex')
plt.title("Survival w.r.t sex of passengers")
```

```python
sns.countplot(data = df, x = 'Survived', hue = 'Pclass')
plt.title("Survival w.r.t class of passenger tickets")
plt.show()
```

* Most of the people from class P = 3 did not survive.
* Most of the P = 1 class passengers survived.
* Passengers of class P = 2 survived and died almost equally

```python
sns.countplot(data = df, x = 'Survived', hue = 'sibsp')
plt.title("Survival w.r.t # siblings / spouses aboard the Titanic")
plt.show()
```

```python
sns.countplot(data = df, x = 'Survived', hue = 'sibsp')
plt.title("Survival w.r.t # siblings / spouses aboard the Titanic")
plt.show()
```

```python
grid = sns.FacetGrid(df, col='Survived')
grid.map(sns.barplot, 'Sex', 'Fare', alpha=.5)
grid.add_legend()
plt.show()
```

```python
sns.set(font_scale=1.15)
plt.figure(figsize=(8,4))
sns.heatmap( df.drop('Passengerid', axis=1).corr(), cmap='RdBu_r', annot=True)
plt.show()
```

```python
df = sns.load_dataset("titanic")
sns.scatterplot(x="age", y="fare", data=df)
plt.title("Age vs Fare in Titanic Dataset")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

visual demonstration of pattern between two continuous values in the titanic dataset.
